## loc / iloc

1. loc: (label-based) se basa en etiquetas "Name", "Age", "Ticket", ...
2. iloc: (integer position-based) se basa en indices numéricos 456, 0, 234, ...

```python
# obtenemos una sola fila usando el indice
df.iloc[78]
```

```python
# obtenemos varias filas usando sus indices
df.iloc[[100, 105]]
```

```python
# Escogiendo filas y columnas específicas

df.loc[[400,401,402,405], ["Sex", "Age", "Survived"]]
```

```python
# Escogiendo filas y columnas específicas

df[400:405].loc[:,["Sex", "Age", "Survived"]]
```

```python
# también podemos filtrar los datos 

df[df["Age"] < 2]
```

```python
df[df["Embarked"] == 'S']
```

```python
# eliminar una columna

del df['Embarked']

df
```

```python
# insertar una columna nueva

df.insert(1, 'TravelWithFamily', df['SibSp'] > 1)
df
```
